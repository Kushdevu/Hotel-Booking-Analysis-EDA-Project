## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Handling Duplicates
df.drop_duplicates(inplace=True)
df.shape
````

````python
#Handling Missing Values
#Imputing Null values with appropriate measures
df['company'].fillna(df['company'].astype('float').mean(), inplace=True)
df['country'].fillna(df['country'].mode().to_string(), inplace=True)
df['agent'].fillna(0, inplace=True)
df['agent'] = df['agent'].round()
df['children'].fillna(0, inplace=True)
````

````python
df.isna().sum()
````

````python
df.dtypes
````

### **Adding some columns.**

````python
#Adding total stay column in dataframe that is (stays_in_week_nights + stays_in_weekend_nights)
df['total_stay'] = df['stays_in_week_nights']+df['stays_in_weekend_nights']

#Adding total people columns that is(adults + children + babies)
df['total_people'] = df['adults'] + df['children'] + df['babies']
````
